```python
import pandas as pd
import os

# Definisi split yang ingin digabung
subjects = [f'SUB{i}' for i in range(1, 13)]
splits = ['train', 'val', 'test']

print("Mulai menggabungkan file per split...\n")

for split in splits:
    print(f"--- MENGGABUNGKAN SPLIT: {split.upper()} ---")
    
    # Nama file target output
    target_filename = f"all_{split}.csv"
    
    # Pindahkan list penampung ke SINI (di luar loop subjek)
    df_list = []
    
    for subject in subjects:
        file_name = f"{subject}_{split}.csv"
        
        # Langsung cek apakah file tersebut ada
        if os.path.exists(file_name):
            df = pd.read_csv(file_name)
            df_list.append(df)
            print(f"  -> Membaca {file_name} ({len(df)} baris)")
        else:
            print(f"  ⚠️ Melewati {file_name} (File tidak ditemukan)")
            
    # Setelah semua subjek dicek dan dikumpulkan, barulah digabungkan
    if df_list:
        # Gabungkan (concat) semua DataFrame dalam list menjadi satu
        all_df = pd.concat(df_list, ignore_index=True)
        
        # Simpan file gabungan
        all_df.to_csv(target_filename, index=False)
        print(f"✅ BERHASIL! Disimpan sebagai '{target_filename}' (Total Akhir: {len(all_df)} baris)\n")
    else:
        print(f"❌ GAGAL: Tidak ada satupun file '{split}' yang ditemukan untuk digabung.\n")

# Preview hasil salah satu file untuk memastikan
if os.path.exists("all_val.csv"):
    print("Preview tabel all_val.csv:")
    df_preview = pd.read_csv("all_val.csv")
    display(df_preview.head())
```
