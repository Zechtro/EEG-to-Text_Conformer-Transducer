# Exploratory Data Analysis - Kalimat (Sentences)

Analisis mendalam tentang data kalimat dari transcript mapping CSV termasuk:
- Analisis keseluruhan (overall analysis)
- Analisis per subjek
- Karakter unik, kata unik, tanda baca, dan angka

```python
import pandas as pd
import os
from collections import Counter
import string
import re

# Get the path to the CSV file
csv_path = "../../../dataset/raw_transcript_mapping.csv"

# Load the data
df = pd.read_csv(csv_path)

print(f"Data dimuat dengan sukses!")
print(f"Total baris: {len(df)}")
print(f"\nKolom: {df.columns.tolist()}")
print(f"\nFirst 5 rows:")
print(df.head())
```

## 1. ANALISIS KESELURUHAN (OVERALL ANALYSIS)

```python
# Combine all sentences
all_sentences = df['sentence'].astype(str).str.cat(sep=' ')

# Basic statistics
total_sentences = len(df)
total_chars_with_space = len(all_sentences)
total_chars_no_space = len(all_sentences.replace(' ', ''))

print("=" * 80)
print("STATISTIK DASAR")
print("=" * 80)
print(f"Total Kalimat: {total_sentences}")
print(f"Total Karakter (dengan spasi): {total_chars_with_space}")
print(f"Total Karakter (tanpa spasi): {total_chars_no_space}")
print(f"Rata-rata Karakter per Kalimat: {total_chars_no_space / total_sentences:.2f}")
```

### Karakter Unik

```python
# Unique characters (only letters and spaces, case insensitive)
all_sentences_lower = all_sentences.lower()
unique_chars = set(c for c in all_sentences_lower if c.isalpha() or c == ' ')
char_counts = Counter(c for c in all_sentences_lower if c.isalpha() or c == ' ')

print(f"Jumlah Karakter Unik (huruf dan spasi, case insensitive): {len(unique_chars)}")
print(f"\nKarakter Unik:")
print(sorted(unique_chars))

print(f"\nFrekuensi Setiap Huruf dan Spasi:")
print("-" * 50)
for char, count in char_counts.most_common():
    display_char = "spasi" if char == ' ' else char
    print(f"  '{display_char}': {count:4d}")
```

### Value Count Kata Unik

```python
# Unique words and their counts
words = all_sentences.lower().split()
unique_words = set(words)
word_counts = Counter(words)

print(f"Jumlah Kata Unik: {len(unique_words)}")
print(f"Total Kata: {len(words)}")
print(f"\nTop 30 Kata Paling Sering:")
print("-" * 50)
for i, (word, count) in enumerate(word_counts.most_common(30), 1):
    print(f"{i:2d}. {word:20s} : {count:4d}")
```

### Tanda Baca

```python
# Punctuation marks
punctuation_in_text = set(c for c in all_sentences if c in string.punctuation)
punct_counts = Counter(c for c in all_sentences if c in string.punctuation)

print(f"Jumlah Tanda Baca Unik: {len(punctuation_in_text)}")
print(f"Tanda Baca yang Ada: {sorted(punctuation_in_text)}")

if punct_counts:
    print(f"\nFrekuensi Tanda Baca:")
    print("-" * 40)
    for punct, count in punct_counts.most_common():
        print(f"  '{punct}': {count}")
else:
    print("\nTidak ada tanda baca dalam data.")
```

### Angka

```python
# Numbers
numbers_in_text = re.findall(r'\d+', all_sentences)
unique_numbers = set(numbers_in_text)
number_counts = Counter(numbers_in_text)

print(f"Ada Angka: {'Ya' if unique_numbers else 'Tidak'}")
if unique_numbers:
    print(f"Angka Unik: {sorted(unique_numbers)}")
    print(f"\nFrekuensi Angka:")
    print("-" * 40)
    for num, count in number_counts.most_common():
        print(f"  {num}: {count}")
else:
    print("\nTidak ada angka dalam data.")
```

## 2. ANALISIS PER SUBJEK (PER SUBJECT ANALYSIS)

```python
# Get unique subjects
subjects = sorted(df['subject'].unique())
print(f"Total Subjek: {len(subjects)}")
print(f"Subjek: {subjects}")
```

```python
# Per-subject analysis summary
subject_summary = []

for subject in subjects:
    subject_df = df[df['subject'] == subject]
    subject_sentences = subject_df['sentence'].astype(str).str.cat(sep=' ')
    
    gender = subject_df['gender'].iloc[0]
    total_kalimat = len(subject_df)
    total_chars = len(subject_sentences.replace(' ', ''))
    avg_chars = total_chars / total_kalimat
    
    subject_summary.append({
        'Subject': subject,
        'Gender': gender,
        'Total Kalimat': total_kalimat,
        'Total Karakter': total_chars,
        'Rata-rata Karakter/Kalimat': avg_chars
    })

summary_df = pd.DataFrame(subject_summary)
print(summary_df.to_string(index=False))
```

```python
# Detailed per-subject analysis
for subject in subjects:
    subject_df = df[df['subject'] == subject]
    subject_sentences = subject_df['sentence'].astype(str).str.cat(sep=' ')
    
    print("\n" + "=" * 80)
    print(f"SUBJEK: {subject}")
    print("=" * 80)
    
    # Basic info
    print(f"Gender: {subject_df['gender'].iloc[0]}")
    print(f"Total Kalimat: {len(subject_df)}")
    print(f"Total Karakter (tanpa spasi): {len(subject_sentences.replace(' ', ''))}")
    print(f"Rata-rata Karakter per Kalimat: {len(subject_sentences.replace(' ', '')) / len(subject_df):.2f}")
    
    # Unique characters (only letters and spaces, case insensitive)
    subject_sentences_lower = subject_sentences.lower()
    subject_unique_chars = set(c for c in subject_sentences_lower if c.isalpha() or c == ' ')
    subject_char_counts = Counter(c for c in subject_sentences_lower if c.isalpha() or c == ' ')
    
    print(f"\nJumlah Karakter Unik (huruf dan spasi, case insensitive): {len(subject_unique_chars)}")
    print(f"Karakter Unik:")
    print(sorted(subject_unique_chars))
    
    print(f"\nFrekuensi Setiap Huruf dan Spasi:")
    print("-" * 50)
    for char, count in subject_char_counts.most_common():
        display_char = "spasi" if char == ' ' else char
        print(f"  '{display_char}': {count:4d}")
    
    # Unique words
    subject_words = subject_sentences.lower().split()
    subject_unique_words = set(subject_words)
    subject_word_counts = Counter(subject_words)
    
    print(f"\nJumlah Kata Unik: {len(subject_unique_words)}")
    print(f"\nTop 15 Kata Paling Sering:")
    print("-" * 50)
    for i, (word, count) in enumerate(subject_word_counts.most_common(15), 1):
        print(f"{i:2d}. {word:20s} : {count:4d}")
    
    # Punctuation
    subject_punctuation = set(c for c in subject_sentences if c in string.punctuation)
    print(f"\nTanda Baca yang Ada: {sorted(subject_punctuation) if subject_punctuation else 'Tidak ada'}")
    
    if subject_punctuation:
        subject_punct_counts = Counter(c for c in subject_sentences if c in string.punctuation)
        print("Frekuensi Tanda Baca:")
        for punct, count in subject_punct_counts.most_common():
            print(f"  '{punct}': {count}")
    
    # Numbers
    subject_numbers = re.findall(r'\d+', subject_sentences)
    subject_unique_numbers = set(subject_numbers)
    print(f"\nAda Angka: {'Ya' if subject_unique_numbers else 'Tidak'}")
    if subject_unique_numbers:
        print(f"Angka Unik: {sorted(subject_unique_numbers)}")
        subject_number_counts = Counter(subject_numbers)
        print("Frekuensi Angka:")
        for num, count in subject_number_counts.most_common():
            print(f"  {num}: {count}")

print("\n" + "=" * 80)
print("ANALISIS SELESAI")
print("=" * 80)
```
