```python
import os, numpy as np, pandas as pd
from IPython.display import display

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '../../../'))
DATA_DIR     = os.path.join(PROJECT_ROOT, 'experiments/ZuCo')
FINAL_DIR    = os.path.join(DATA_DIR, 'final_comparison')
os.makedirs(FINAL_DIR, exist_ok=True)

metrics = ['cer','WER','BLEU-1','BLEU-2','BLEU-3','BLEU-4',
           'ROUGE-1-P','ROUGE-1-R','ROUGE-1-F','BertScore']

experiments = [
    {'exp_id':'5', 'decoder':'LSTM',       'label':'Hilbert + Char LSTM (v6_1)',
     'file':'complete_metrics_ZuCo_ZGW_NR_hilbert_fast_test_predictions_6_1.csv'},
    {'exp_id':'6', 'decoder':'GPT-2 Frozen','label':'Hilbert + GPT-2 Frozen (v10_1)',
     'file':'complete_metrics_ZuCo_ZGW_NR_hilbert_GPT2_fast_test_predictions_10_1.csv'},
]

rows = []
for exp in experiments:
    fp  = os.path.join(DATA_DIR, exp['file'])
    row = {'Exp': exp['exp_id'], 'Decoder': exp['decoder'], 'Label': exp['label']}
    if os.path.exists(fp):
        df = pd.read_csv(fp)
        for m in metrics:
            row[m] = round(df[m].mean(), 4) if m in df.columns else np.nan
    else:
        print(f"WARNING: {exp['file']} not found")
        for m in metrics: row[m] = np.nan
    rows.append(row)

df_out = pd.DataFrame(rows)[['Exp','Label','Decoder'] + metrics]
print('='*100)
print('ZuCo | HILBERT | TEST PREDICTIONS (ZGW, NR)')
print('='*100)
display(df_out)
out = os.path.join(FINAL_DIR, 'ZuCo_hilbert_evaluation_summary.csv')
df_out.to_csv(out, index=False)
print(f'Saved: {out}')
```
