```python
import os
import pandas as pd
from IPython.display import display

# Ubah nilai ini untuk mengganti subjek (contoh: 'SUB1', 'SUB2', dst.)
SUBJECT = 'SUB1'

# Ubah nilai ini untuk mengganti metrik target (contoh: 'cer', 'WER', 'BLEU-1', dll.)
TARGET_METRIC = 'cer'

PERCENTILES = [0, 10, 25, 35, 50, 65, 75, 85, 95, 100]

script_dir = os.getcwd()

file_templates = [
    "complete_metrics_{subject}_eq_3_0_fixed_hilbert_test_predictions_6_1.csv",
    "complete_metrics_{subject}_eq_3_0_hilbert_test_predictions_10_1_IndoGPT.csv",
    "complete_metrics_{subject}_eq_3_0_log-mel_test_predictions_6_1.csv",
    "complete_metrics_{subject}_eq_3_0_logmel_test_predictions_10_1_IndoGPT.csv",
    
    "complete_metrics_NOISE_BASELINE_{subject}_eq_3_0_fixed_hilbert_test_predictions_6_1.csv",
    "complete_metrics_NOISE_BASELINE_{subject}_eq_3_0_hilbert_test_predictions_10_1_IndoGPT.csv",
    "complete_metrics_NOISE_BASELINE_{subject}_eq_3_0_log-mel_test_predictions_6_1.csv",
    "complete_metrics_NOISE_BASELINE_{subject}_eq_3_0_logmel_test_predictions_10_1_IndoGPT.csv"
]

print(f"Menampilkan sampel prediksi pada persentil {TARGET_METRIC.upper()} untuk subjek: {SUBJECT}\n")

for template in file_templates:
    filename = template.format(subject=SUBJECT)
    filepath = os.path.join(script_dir, filename)
    
    print("=" * 110)
    print(f"📄 File: {filename}")
    print("=" * 110)
    
    if os.path.exists(filepath):
        try:
            df = pd.read_csv(filepath)
            
            req_cols = [TARGET_METRIC, 'sentence', 'prediction']
            if all(col in df.columns for col in req_cols):
                
                df_sorted = df.sort_values(by=TARGET_METRIC).reset_index(drop=True)
                
                results = []
                n_rows = len(df_sorted)
                
                if n_rows > 0:
                    for p in PERCENTILES:
                        # Rumus: (p / 100) * (total_baris - 1)
                        idx = int((p / 100.0) * (n_rows - 1))
                        row = df_sorted.iloc[idx]
                        
                        results.append({
                            'persentil': f"{p}",
                            TARGET_METRIC: round(row[TARGET_METRIC], 4),
                            'sentence': row['sentence'],
                            'prediction': row['prediction']
                        })
                    
                    df_result = pd.DataFrame(results)
                    display(df_result)
                else:
                    print("⚠️ File CSV kosong (tidak ada baris).")
            else:
                missing = [col for col in req_cols if col not in df.columns]
                print(f"⚠️ Peringatan: Kolom {missing} tidak ditemukan di dalam file ini.")
                
        except Exception as e:
            print(f"❌ Gagal membaca file: {e}")
    else:
        print(f"❌ File belum ditemukan di: {filepath}")
        
    print("\n")
```
