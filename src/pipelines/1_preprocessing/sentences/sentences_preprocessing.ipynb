# Sentences Preprocessing

Data cleaning untuk kalimat/sentences:
- Remove punctuation (keep spaces)
- Lowercase

```python
import pandas as pd
import os
import re
import string

# Get the project root directory
notebook_dir = os.path.dirname(os.path.abspath('__file__'))
project_root = os.path.abspath(os.path.join(notebook_dir, '../../../../'))

# Define paths
raw_csv_path = os.path.join(project_root, 'dataset/raw_transcript_mapping.csv')
cleaned_csv_path = os.path.join(project_root, 'dataset/cleaned_transcript_mapping.csv')

print(f"Raw CSV Path: {raw_csv_path}")
print(f"Cleaned CSV Path: {cleaned_csv_path}")

# Load the raw data
df = pd.read_csv(raw_csv_path)
print(f"\nData loaded successfully!")
print(f"Shape: {df.shape}")
print(f"\nFirst 5 rows:")
print(df.head())
```

## Data Cleaning Function

```python
def clean_sentence(sentence):
    """
    Clean a sentence by:
    1. Converting to lowercase
    2. Removing punctuation (keep spaces)
    """
    # Convert to lowercase
    sentence = str(sentence).lower()
    
    # Remove punctuation, keeping spaces
    sentence = ''.join(char if char not in string.punctuation else ' ' if char in '.,;:!?\'"()' else '' 
                       for char in sentence)
    
    # Remove extra spaces
    sentence = ' '.join(sentence.split())
    
    return sentence

# Test the function
test_sentence = df['sentence'].iloc[0]
print(f"Original: {test_sentence}")
print(f"Cleaned: {clean_sentence(test_sentence)}")
```

## Apply Cleaning

```python
# Create a copy of the dataframe
df_cleaned = df.copy()

# Apply cleaning to the sentence column
df_cleaned['sentence'] = df_cleaned['sentence'].apply(clean_sentence)

print("Cleaning completed!")
print(f"\nFirst 10 rows (before and after):")
print(pd.DataFrame({
    'Original': df['sentence'].head(10).values,
    'Cleaned': df_cleaned['sentence'].head(10).values
}))
```

## Save Cleaned Data

```python
# Save the cleaned data
df_cleaned.to_csv(cleaned_csv_path, index=False)

print(f"Cleaned data saved to: {cleaned_csv_path}")
print(f"Total rows saved: {len(df_cleaned)}")
print(f"\nCleaned data preview:")
print(df_cleaned.head(10))
```

## Summary Statistics

```python
print("=" * 80)
print("PREPROCESSING SUMMARY")
print("=" * 80)

# Compare original and cleaned
avg_original_length = df['sentence'].str.len().mean()
avg_cleaned_length = df_cleaned['sentence'].str.len().mean()

print(f"\nOriginal Data:")
print(f"  Total sentences: {len(df)}")
print(f"  Average sentence length: {avg_original_length:.2f}")
print(f"  Min length: {df['sentence'].str.len().min()}")
print(f"  Max length: {df['sentence'].str.len().max()}")

print(f"\nCleaned Data:")
print(f"  Total sentences: {len(df_cleaned)}")
print(f"  Average sentence length: {avg_cleaned_length:.2f}")
print(f"  Min length: {df_cleaned['sentence'].str.len().min()}")
print(f"  Max length: {df_cleaned['sentence'].str.len().max()}")

print(f"\nAverage reduction in length: {((avg_original_length - avg_cleaned_length) / avg_original_length * 100):.2f}%")
print(f"\nFile saved successfully to: {cleaned_csv_path}")
```
